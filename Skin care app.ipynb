```python
import random

def get_skin_type():
    print("Let's find out your skin type and recommend a skincare routine!")
    print("Answer the following questions to determine your skin type:")

    # Question 1
    oiliness = input("Does your skin feel oily by the end of the day? (yes/no): ").lower()

    # Question 2
    dryness = input("Does your skin feel tight or dry, especially after washing? (yes/no): ").lower()

    # Question 3
    breakouts = input("Do you experience frequent breakouts? (yes/no): ").lower()

    # Determine skin type
    if oiliness == "yes" and dryness == "no" and breakouts == "yes":
        skin_type = "Oily and Acne-Prone"
    elif oiliness == "no" and dryness == "yes" and breakouts == "no":
        skin_type = "Dry and Sensitive"
    elif oiliness == "yes" and dryness == "yes" and breakouts == "no":
        skin_type = "Combination"
    elif oiliness == "no" and dryness == "no" and breakouts == "no":
        skin_type = "Normal"
    else:
        skin_type = "Mixed Skin"

    return skin_type

def get_skincare_routine(skin_type):
    # Based on skin type, recommend a skincare routine
    if skin_type == "Oily and Acne-Prone":
        routine = """
        1. Cleanser🧼: Gel-based, oil-free, and non-comedogenic.
        2. Exfoliate🧖‍♀️: 2-3 times a week with a BHA (Salicylic Acid) exfoliant.
        3. Moisturizer💆‍♀️: Lightweight, oil-free.
        4. Sunscreen🧴: Broad-spectrum, oil-free, and mattifying.
        """
    elif skin_type == "Dry and Sensitive":
        routine = """
        1. Cleanser🧼: Hydrating and gentle cream-based cleanser.
        2. Exfoliate🧖‍♀️: Avoid exfoliation or use a mild exfoliant.
        3. Moisturizer💆‍♀️: Rich, nourishing cream with ceramides.
        4. Sunscreen🧴: Hydrating sunscreen with added moisturizers.
        """
    elif skin_type == "Combination":
        routine = """
        1. Cleanser🧼: Gentle foaming or gel cleanser.
        2. Exfoliate🧖‍♀️: Exfoliate 1-2 times a week.
        3. Moisturizer💆‍♀️: Lightweight gel for oily areas and richer cream for dry areas.
        4. Sunscreen🧴: Non-comedogenic, light formula.
        """
    elif skin_type == "Normal":
        routine = """
        1. Cleanser🧼: Mild, hydrating cleanser.
        2. Exfoliate🧖‍♀️: Exfoliate 1-2 times a week.
        3. Moisturizer💆‍♀️: Light lotion or cream.
        4. Sunscreen🧴: Hydrating sunscreen with broad-spectrum protection.
        """
    else:
        routine = """
        1. Cleanser🧼: Gentle hydrating cleanser.
        2. Exfoliate🧖‍♀️: Use a mild exfoliant twice a week.
        3. Moisturizer💆‍♀️: A balanced moisturizer for all skin types.
        4. Sunscreen🧴: Lightweight, broad-spectrum sunscreen.
        """

    return routine

def main():
    # Get skin type from the user
    skin_type = get_skin_type()

    print(f"\nYour skin type is: {skin_type}")


    routine = get_skincare_routine(skin_type)
    print("\nRecommended Skincare Routine:")
    print(routine)

# Run the app
if __name__ == "__main__":
    main()
```
